```python

import numpy as np

# 데이터 생성: 2차원 데이터를 예시로 사용
np.random.seed(0)
data = np.random.randn(100, 2)

# 1. 데이터의 평균 계산 및 평균 중심화
mean = np.mean(data, axis=0)
centered_data = data - mean

# 2. 공분산 행렬 계산
cov_matrix = np.cov(centered_data, rowvar=False)

# 3. 공분산 행렬의 고유값과 고유벡터 계산
eig_vals, eig_vecs = np.linalg.eigh(cov_matrix)

# 4. 고유값을 기준으로 고유벡터 정렬 (내림차순)
sorted_indices = np.argsort(eig_vals)[::-1]
sorted_eig_vecs = eig_vecs[:, sorted_indices]

# 5. 데이터를 주성분에 투영
pca_result = np.dot(centered_data, sorted_eig_vecs)

# 결과 출력
print("원본 데이터의 주성분:\n", sorted_eig_vecs)
print("PCA 적용 후 데이터:\n", pca_result)
```
