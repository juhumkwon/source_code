```python
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np

# 잠재 변수 z의 차원
latent_dim = 2

# 인코더 클래스 정의
class Encoder(Model):
    def __init__(self, latent_dim):
        super(Encoder, self).__init__()
        # 첫 번째 은닉층 (입력 크기 784 -> 256)
        self.dense1 = layers.Dense(256, activation='relu')
        # 두 번째 은닉층 (256 -> 128)
        self.dense2 = layers.Dense(128, activation='relu')
        # 평균 출력층 (128 -> latent_dim)
        self.mean = layers.Dense(latent_dim)
        # 로그-분산 출력층 (128 -> latent_dim)
        self.log_var = layers.Dense(latent_dim)

    def call(self, x):
        # 은닉층 통과
        x = self.dense1(x)
        x = self.dense2(x)
        # 평균과 로그-분산 계산
        mu = self.mean(x)
        log_var = self.log_var(x)
        return mu, log_var

# 리파라미터라이제이션 트릭 (평균과 로그-분산으로부터 z 샘플링)
def reparameterization(mu, log_var):
    # 표준 정규분포에서 샘플링한 값
    epsilon = tf.random.normal(shape=mu.shape)
    # 표준편차는 exp(log(𝜎²)) = sqrt(𝜎²)
    sigma = tf.exp(0.5 * log_var)
    # z 샘플링
    z = mu + sigma * epsilon
    return z

# 입력 데이터 예시 (배치 크기 64, 28x28 이미지 크기 -> 784 차원)
batch_size = 64
input_dim = 784  # 28x28 이미지를 1D 벡터로 변환
x_input = np.random.rand(batch_size, input_dim).astype(np.float32)

# 인코더 인스턴스 생성
encoder = Encoder(latent_dim)

# 입력 데이터를 인코더에 전달하여 평균과 로그-분산 계산
mu, log_var = encoder(x_input)

# 리파라미터라이제이션을 통해 잠재 변수 z 샘플링
z = reparameterization(mu, log_var)

# 결과 출력
print("평균 (mu):", mu.numpy())
print("로그-분산 (log_var):", log_var.numpy())
print("잠재 변수 z 샘플링 결과:", z.numpy())
```
