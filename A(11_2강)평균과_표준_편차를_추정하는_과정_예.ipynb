```python
import tensorflow as tf
from tensorflow.keras import layers, Model
import numpy as np

# ì ì¬ ë³€ìˆ˜ zì˜ ì°¨ì›
latent_dim = 2

# ì¸ì½”ë” í´ë˜ìŠ¤ ì •ì˜
class Encoder(Model):
    def __init__(self, latent_dim):
        super(Encoder, self).__init__()
        # ì²« ë²ˆì§¸ ì€ë‹‰ì¸µ (ì…ë ¥ í¬ê¸° 784 -> 256)
        self.dense1 = layers.Dense(256, activation='relu')
        # ë‘ ë²ˆì§¸ ì€ë‹‰ì¸µ (256 -> 128)
        self.dense2 = layers.Dense(128, activation='relu')
        # í‰ê·  ì¶œë ¥ì¸µ (128 -> latent_dim)
        self.mean = layers.Dense(latent_dim)
        # ë¡œê·¸-ë¶„ì‚° ì¶œë ¥ì¸µ (128 -> latent_dim)
        self.log_var = layers.Dense(latent_dim)

    def call(self, x):
        # ì€ë‹‰ì¸µ í†µê³¼
        x = self.dense1(x)
        x = self.dense2(x)
        # í‰ê· ê³¼ ë¡œê·¸-ë¶„ì‚° ê³„ì‚°
        mu = self.mean(x)
        log_var = self.log_var(x)
        return mu, log_var

# ë¦¬íŒŒë¼ë¯¸í„°ë¼ì´ì œì´ì…˜ íŠ¸ë¦­ (í‰ê· ê³¼ ë¡œê·¸-ë¶„ì‚°ìœ¼ë¡œë¶€í„° z ìƒ˜í”Œë§)
def reparameterization(mu, log_var):
    # í‘œì¤€ ì •ê·œë¶„í¬ì—ì„œ ìƒ˜í”Œë§í•œ ê°’
    epsilon = tf.random.normal(shape=mu.shape)
    # í‘œì¤€í¸ì°¨ëŠ” exp(log(ğœÂ²)) = sqrt(ğœÂ²)
    sigma = tf.exp(0.5 * log_var)
    # z ìƒ˜í”Œë§
    z = mu + sigma * epsilon
    return z

# ì…ë ¥ ë°ì´í„° ì˜ˆì‹œ (ë°°ì¹˜ í¬ê¸° 64, 28x28 ì´ë¯¸ì§€ í¬ê¸° -> 784 ì°¨ì›)
batch_size = 64
input_dim = 784  # 28x28 ì´ë¯¸ì§€ë¥¼ 1D ë²¡í„°ë¡œ ë³€í™˜
x_input = np.random.rand(batch_size, input_dim).astype(np.float32)

# ì¸ì½”ë” ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
encoder = Encoder(latent_dim)

# ì…ë ¥ ë°ì´í„°ë¥¼ ì¸ì½”ë”ì— ì „ë‹¬í•˜ì—¬ í‰ê· ê³¼ ë¡œê·¸-ë¶„ì‚° ê³„ì‚°
mu, log_var = encoder(x_input)

# ë¦¬íŒŒë¼ë¯¸í„°ë¼ì´ì œì´ì…˜ì„ í†µí•´ ì ì¬ ë³€ìˆ˜ z ìƒ˜í”Œë§
z = reparameterization(mu, log_var)

# ê²°ê³¼ ì¶œë ¥
print("í‰ê·  (mu):", mu.numpy())
print("ë¡œê·¸-ë¶„ì‚° (log_var):", log_var.numpy())
print("ì ì¬ ë³€ìˆ˜ z ìƒ˜í”Œë§ ê²°ê³¼:", z.numpy())
```
