```python
from sklearn import datasets
from sklearn import svm
import matplotlib.pyplot as plt

#   (:  )
iris = datasets.load_iris()
X = iris.data[:, :2]  #    
y = iris.target

# SVM   (RBF  )
model = svm.SVC(kernel='rbf', gamma='scale')  # gamma RBF  
model.fit(X, y)

#  
"""
plt.scatter:
Matplotlib pyplot   ,    .
- X[:, 0]:   X    . X 2 ,     ,    .
           X[:, 0]      (  ) .
- X[:, 1]:   X    . , X[:, 1]      (  ) .

- c=y:    . y       .      .
- cmap='viridis':  (color map) . 'viridis' Matplotlib     ,         .
- s=30:    .  s=30    30 .
"""
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=30)
plt.title("SVM with RBF Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print(X[:, 0])
print("\n")
print(X[:, 1])
```
