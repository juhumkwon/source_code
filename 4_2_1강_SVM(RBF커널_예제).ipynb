```python
from sklearn import datasets
from sklearn import svm
import matplotlib.pyplot as plt

# 데이터셋 로드 (예: 아이리스 데이터)
iris = datasets.load_iris()
X = iris.data[:, :2]  # 두 개의 특성 사용
y = iris.target

# SVM 모델 생성 (RBF 커널 사용)
model = svm.SVC(kernel='rbf', gamma='scale')  # gamma는 RBF 커널의 파라미터
model.fit(X, y)

# 결과 시각화
"""
plt.scatter:
Matplotlib의 pyplot 모듈에서 제공하는 함수로, 산점도를 그릴 때 사용됩니다.
- X[:, 0]: 데이터 배열 X의 첫 번째 특성을 선택합니다. X는 2차원 배열이며, 각 행은 데이터 포인트를 나타내고, 각 열은 특성을 나타냅니다.
           X[:, 0]는 모든 데이터 포인트의 첫 번째 특성(첫 번째 열)을 선택합니다.
- X[:, 1]: 데이터 배열 X의 두 번째 특성을 선택합니다. 마찬가지로, X[:, 1]은 모든 데이터 포인트의 두 번째 특성(두 번째 열)을 선택합니다.

- c=y: 산점도의 점 색상을 지정합니다. y는 각 데이터 포인트의 클래스 레이블을 나타내는 배열입니다. 각 클래스는 서로 다른 색으로 표시됩니다.
- cmap='viridis': 색상 맵(color map)을 지정합니다. 'viridis'는 Matplotlib에서 제공하는 색상 맵 중 하나로, 시각적으로 보기 좋고 색상의 연속성을 잘 나타내는 특성이 있습니다.
- s=30: 각 점의 크기를 지정합니다. 여기서 s=30은 산점도의 점 크기를 30으로 설정합니다.
"""
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=30)
plt.title("SVM with RBF Kernel")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

print(X[:, 0])
print("\n")
print(X[:, 1])
```
