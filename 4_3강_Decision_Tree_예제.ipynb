```python
#출처 : https://github.com/gilbutITbook/080263/tree/master/chap3/data/titanic
from google.colab import files
file_uploaded=files.upload()

import pandas as pd
df = pd.read_csv('train.csv', index_col='PassengerId')
print(df.head())

#데이터 전처리
#데이터 분석에 필요한 컬럼만 추출
df = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived']]
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = df.dropna() #null 값이 있는 열 삭제

"""
df.drop('Survived', axis=1): 이 메서드는 데이터프레임 df에서 'Survived'라는 열을 제거합니다.
axis=1은 열을 기준으로 작업을 수행하겠다는 의미입니다. (행을 기준으로 작업을 수행하려면 axis=0을 사용합니다.)
"""
X = df.drop('Survived', axis=1)
#print("\n X=", X)
y = df['Survived']
#print("\n y=", y)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

from sklearn import tree
model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)
y_predict = model.predict(X_test)

from sklearn.metrics import accuracy_score

# 검증 데이터에 대한 예측 결과
print("accuracy_score(y_test, y_predict", accuracy_score(y_test, y_predict))

#혼동 행렬을 이용한 성능 측정
from sklearn.metrics import confusion_matrix

pd.DataFrame(
	confusion_matrix(y_test, y_predict),
	columns = ['Predicted Not Survival', 'Predicted Survival'],
    index = ['True Not Survival', 'True Survival']
)
```

```python
"""
DataFrame()의 주요 개념
2차원 자료구조: 행(row)과 열(column)로 구성되어 있으며, 각 열은 서로 다른 데이터 타입을 가질 수 있습니다.
인덱스(index): 각 행의 고유한 식별자 역할을 하며, 기본적으로 0부터 시작하는 숫자가 할당되지만, 사용자가 지정할 수도 있습니다.
열 이름(columns): 각 열에 고유한 이름이 있으며, 이를 통해 특정 데이터를 쉽게 조회하고 조작할 수 있습니다.
"""
import pandas as pd

# 딕셔너리 데이터를 사용하여 DataFrame 생성
data = {
    'Column1': [1, 2, 3],
    'Column2': ['A', 'B', 'C']
}

df = pd.DataFrame(data)

# DataFrame 출력
print(df)
```

```python
#dropna() 예제
"""
df = df.dropna()는 pandas DataFrame에서 **결측값(NA 또는 NaN, 즉 null 값)**이 포함된 행을 삭제하는 함수입니다.
이 코드는 df에서 하나라도 결측값이 있는 행을 제거한 후, 그 결과를 다시 df에 저장하는 역할을 합니다.
"""
import pandas as pd
import numpy as np

# 데이터 생성 (결측값 포함)
data = {
    'Pclass': [1, 3, 2, np.nan],
    'Sex': ['male', 'female', np.nan, 'female'],
    'Age': [22, np.nan, 58, 35],
    'Fare': [7.25, 71.83, np.nan, 8.05]
}

# DataFrame 생성
df = pd.DataFrame(data)

# 원본 DataFrame 출력
print("Before dropna():")
print(df)

# 결측값이 포함된 행을 삭제
#df = df.dropna()

#how='all': 모든 값이 결측값일 때만 해당 행을 삭제합니다.
#df=df.dropna(how='all')

#axis=1: 결측값이 있는 열을 삭제합니다.
df=df.dropna(axis=1)

#subset=['column1', 'column2']: 특정 열에서만 결측값을 확인하고 삭제합니다.
#df=df.dropna(subset=['Age', 'Fare'])

# 결측값 제거 후 DataFrame 출력
print("\nAfter dropna():")
print(df)
```

```python
"""
혼동 행렬(confusion matrix)을 pandas DataFrame 형식으로 생성하는 코드입니다.
혼동 행렬은 분류 모델의 성능을 평가하는 데 유용합니다.
각 부분에 대한 설명은 다음과 같습니다:
confusion_matrix(y_test, y_predict):

sklearn.metrics의 함수로, 혼동 행렬을 계산합니다.
y_test는 테스트 세트의 실제 레이블을 포함하고,
y_predict는 모델의 예측 레이블을 포함합니다.
결과는 2x2 행렬로, 참 긍정(true positives), 참 부정(true negatives), 거짓 긍정(false positives), 거짓 부정(false negatives)의 수를 보여줍니다.
pd.DataFrame(conf_matrix, columns = [...], index = [...]):
예)                     예측: 생존하지 않음      예측: 생존
 실제: 생존하지 않음    [[50,                    10],
 실제: 생존)             [5,                     35]]

실제: 생존하지 않음 (50): 모델이 "생존하지 않음"으로 예측했을 때, 실제로도 "생존하지 않음"인 경우.
예측: 생존 (10): 모델이 "생존"으로 예측했을 때, 실제로는 "생존하지 않음"인 경우 (이것은 오차입니다).
실제: 생존 (35): 모델이 "생존"으로 예측했을 때, 실제로도 "생존"인 경우.
예측: 생존하지 않음 (5): 모델이 "생존하지 않음"으로 예측했을 때, 실제로는 "생존"인 경우 (이것도 오차입니다).

혼동 행렬(NumPy 배열)을 pandas DataFrame으로 변환하여 가독성을 높입니다.
columns: 혼동 행렬의 열 레이블로, 예측된 클래스들을 나타냅니다.
index: 혼동 행렬의 행 레이블로, 실제 클래스들을 나타냅니다.
"""
import pandas as pd
from sklearn.metrics import confusion_matrix

# 혼동 행렬 생성
conf_matrix = confusion_matrix(y_test, y_predict)

# 혼동 행렬을 DataFrame으로 변환
conf_matrix_df = pd.DataFrame(
    conf_matrix,
    columns = ['예측: 생존하지 않음', '예측: 생존'],
    index = ['실제: 생존하지 않음', '실제: 생존']
)
print(conf_matrix_df)
```

```python
from sklearn.metrics import accuracy_score

# 실제 레이블 (test set)
y_test = [0, 1, 0, 1, 0, 1]

# 모델이 예측한 레이블
y_predict = [0, 1, 0, 0, 0, 1]

# 정확도 계산
accuracy = accuracy_score(y_test, y_predict)
print(f"Accuracy: {accuracy}")
```
