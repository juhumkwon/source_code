```python
# [출처] 밀도 기반 군집 분석(DBSCAN) 알고리즘을 이용한 군집 분석 샘플 소스 (Python)|작성자 맑은 하늘

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

# 데이터 생성 (초승달 모양 데이터셋)
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# DBSCAN 모델 생성 및 학습
dbscan = DBSCAN(eps=0.2, min_samples=5)  # eps: 0.2, min_samples: 5 설정
labels = dbscan.fit_predict(X)

# 시각화
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.title('DBSCAN Clustering')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```

```python
#[출처] 밀도 기반 군집 분석(DBSCAN)과 주성분 분석(PCA)을 이용한 군집 분석 절차 및 예제|작성자 맑은 하늘

import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN

# 데이터 로드
iris = load_iris()
X = iris.data

# PCA를 이용한 차원 축소
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# DBSCAN 모델 생성 및 학습 (eps=0.3, min_samples=5)
dbscan = DBSCAN(eps=0.3, min_samples=5)

labels = dbscan.fit_predict(X_pca)

# 시각화
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
plt.title('DBSCAN Clustering on Iris Dataset (PCA-reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
