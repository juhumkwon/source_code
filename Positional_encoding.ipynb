```python

import numpy as np
import matplotlib.pyplot as plt

# Positional Encoding 계산 함수
def positional_encoding(pos, d_model):
    PE = np.zeros((pos, d_model))
    for i in range(d_model):
        if i % 2 == 0:  # 짝수 인덱스에 sin 적용
            PE[:, i] = np.sin(np.arange(pos) / np.power(10000, (2 * i) / d_model))
        else:  # 홀수 인덱스에 cos 적용
            PE[:, i] = np.cos(np.arange(pos) / np.power(10000, (2 * i) / d_model))
    return PE

# 10개의 단어(토큰), 4차원 벡터로 Positional Encoding 생성
PE_example = positional_encoding(pos=10, d_model=4)

# 결과 출력 (위치별 인코딩 값)
print("Positional Encoding (10개 단어, 4차원 벡터):")
print(PE_example)

# 시각화
plt.figure(figsize=(8, 4))
plt.plot(PE_example[:, 0], label="Dim 0 (sin)")
plt.plot(PE_example[:, 1], label="Dim 1 (cos)")
plt.plot(PE_example[:, 2], label="Dim 2 (sin)")
plt.plot(PE_example[:, 3], label="Dim 3 (cos)")
plt.legend()
plt.xlabel("Word Position")
plt.ylabel("Encoding Value")
plt.title("Positional Encoding Values")
plt.show()
```
