```python
#출처:https://datascienceschool.net/03%20machine%20learning/16.02%20K-%ED%8F%89%EA%B7%A0%20%EA%B5%B0%EC%A7%91%ED%99%94.html
import tensorflow as tf
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt # Import the plotting library

#make_blobs: 이 함수는 지정된 개수의 클러스터를 가진 가우시안 분포의 점들을 생성합니다.
#각 클러스터는 랜덤하게 위치한 중심을 가지며, 샘플들은 이 중심 주위에 흩어지게 됩니다.

X, _ = make_blobs(n_samples=20, random_state=4)


"""
    - n_clusters=2: 데이터를 두 개의 클러스터로 나눕니다.
      K-평균 알고리즘은 데이터를 n_clusters개의 클러스터로 분할합니다.

    - init="random": 클러스터 중심(centroid)을 초기화하는 방법을 설정합니다.
      여기서는 "random"으로 설정되어 있어, 초기 중심을 무작위로 선택합니다.
      K-평균 알고리즘은 초기 중심값에 따라 결과가 달라질 수 있기 때문에 초기화 방법이 중요합니다.

    - n_init=1: 다른 초기 중심값을 사용하여 K-평균 알고리즘을 실행하는 횟수입니다.
      여기서는 1번 실행합니다. 일반적으로 n_init 값을 크게 설정하여 여러 번 실행한 후 가장 좋은 결과를 선택하지만,
      여기서는 한 번만 실행합니다.

    - max_iter=n: 알고리즘의 최대 반복 횟수입니다.
      n은 변수로, 알고리즘이 수렴할 때까지 반복하지만, 최대 n번 반복합니다.
      수렴하지 않으면 n번 반복 후 종료됩니다.

    - random_state=6: 난수 생성 시드를 고정하여 재현 가능한 결과를 얻습니다.
      같은 random_state 값을 사용하면 매번 같은 초기 클러스터 중심이 선택됩니다.

    - .fit(X): 주어진 데이터 X에 대해 K-평균 군집화를 실행하여 모델을 학습시킵니다. X는 데이터셋입니다.

    - model.cluster_centers_는 사이킷런(Scikit-learn) 라이브러리에서 K-평균 클러스터링(K-means clustering)
      알고리즘을 사용했을 때, 각 클러스터의 중심점을 나타내는 속성입니다.

 """

def plot_KMeans(n):
    model = KMeans(n_clusters=2, init="random", n_init=1, max_iter=n, random_state=6).fit(X)
    c0, c1 = model.cluster_centers_

    #클러스터 0에 속하는 x축 좌표
    print(X[model.labels_ == 0, 0])
    #클러스터 0에 속하는 y축 좌표
    print(X[model.labels_ == 0, 1])

    #클러스터 1에 속하는 x축 좌표
    print(X[model.labels_ == 1, 0])
    #클러스터 1에 속하는 y축 좌표
    print(X[model.labels_ == 1, 1])

    plt.scatter(X[model.labels_ == 0, 0], X[model.labels_ == 0, 1], marker='v', facecolor='r', edgecolors='k')
    plt.scatter(X[model.labels_ == 1, 0], X[model.labels_ == 1, 1], marker='^', facecolor='y', edgecolors='k')
    plt.scatter(c0[0], c0[1], marker='v', c="r", s=200)
    plt.scatter(c1[0], c1[1], marker='^', c="y", s=200)
    plt.grid(False)
    plt.title("반복횟수={}, 관성={:5.2f}".format(n, -model.score(X)))



plt.figure(figsize=(8, 8))
plt.subplot(321)
plot_KMeans(1)
plt.subplot(322)
plot_KMeans(2)
plt.subplot(323)
plot_KMeans(3)
plt.subplot(324)
plot_KMeans(4)
plt.tight_layout()
plt.show()

```

```python
from sklearn.cluster import MiniBatchKMeans
from sklearn.datasets import make_blobs
import numpy as np

X, y = make_blobs(n_samples=150000, cluster_std=[1.0, 2.5, 0.5], random_state=170)

%time
model1 = KMeans(n_clusters=3).fit(X)

%time
model2 = MiniBatchKMeans(n_clusters=3, batch_size=1000, compute_labels=True).fit(X)

idx = np.random.randint(150000, size=300)
plt.subplot(121)
plt.scatter(X[idx, 0], X[idx, 1], c=model1.labels_[idx])
plt.title("K-평균 군집화")
plt.subplot(122)
plt.scatter(X[idx, 0], X[idx, 1], c=model2.labels_[idx])
plt.title("미니배치 K-평균 군집화")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np

# 샘플 데이터 생성
X = np.array([[1, 2], [1, 4], [1, 0],
              [4, 2], [4, 4], [4, 0]])

# KMeans 모델 생성 및 학습
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)

# 클러스터 0에 속한 데이터 포인트 시각화
plt.scatter(X[kmeans.labels_ == 0, 0], X[kmeans.labels_ == 0, 1], marker='v', facecolor='r', edgecolors='k', label='Cluster 0')

# 클러스터 1에 속한 데이터 포인트 시각화
plt.scatter(X[kmeans.labels_ == 1, 0], X[kmeans.labels_ == 1, 1], marker='o', facecolor='b', edgecolors='k', label='Cluster 1')

# 클러스터 중심 시각화
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='k', label='Centroids')

# 범례 추가 및 그래프 출력
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

"""
X[:, 0]: X축 데이터, 즉 데이터 포인트의 첫 번째 특징을 나타냅니다.

X[:, 1]: Y축 데이터, 즉 데이터 포인트의 두 번째 특징을 나타냅니다.

c=y: 데이터 포인트의 색상을 클러스터 레이블(y)에 따라 설정합니다.
클러스터 레이블이 다르면 서로 다른 색상으로 표시됩니다.

cmap='viridis': 색상 맵을 지정합니다. viridis는 연속적인 색상 맵으로,
클러스터 레이블에 따라 점의 색상을 설정합니다.
색상 맵은 데이터의 연속적인 값을 시각적으로 나타내는 데 유용합니다.

s=1: 각 데이터 포인트의 크기를 설정합니다.
여기서는 점의 크기를 1로 설정하여 많은 데이터 포인트를 시각화할 때 가독성을 높이려는 경우 유용합니다.
"""

# 데이터 생성
X, y = make_blobs(n_samples=150000,
                  cluster_std=[1.0, 2.5, 0.5],
                  random_state=170)

# 생성된 데이터 시각화
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=1)
plt.title('Scatter Plot of Generated Blobs with Different Cluster Std')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# 데이터 생성
X, y = make_blobs(n_samples=150000,
                  cluster_std=[1.0, 2.5, 0.5],
                  random_state=170)

# 데이터 시각화
plt.figure(figsize=(10, 8))  # 그래프 크기 조정
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', s=1, alpha=0.5)
plt.title('Scatter Plot of Generated Blobs with Different Cluster Std')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.colorbar(label='Cluster Label')  # 색상 바 추가
plt.show()
```
